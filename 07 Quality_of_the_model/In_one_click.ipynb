### Построим графики для каждого признака

Создадим функцию для подсчёта доли выбросов в количественных столбцах всех датасетов:

```python
def outliers(dataset, column):
    q1 = dataset[column].quantile(0.25) # первый квартиль
    q3 = dataset[column].quantile(0.75)  # третий квартиль
    iqr = q3 - q1  # межквартильный размах
    quant1 = q1 - 1.5*iqr  # нижняя граница выбросов
    quant3 = q3 + 1.5*iqr  # верхняя граница выбросов
    total = dataset.shape[0]  # общее количество строк в датафрейме
    emission = dataset[(dataset[column] < quant1) | \
     (dataset[column] > quant3)].shape[0]
    # количество выбросов
    emission_quant = (emission / total) * 100  # доля выбросов в столбце
    return np.round((emission_quant), decimals=2)
    # возвращаем долю выбросов в столбце

# Столбцы с выбросами
dataset = market_file
column_list = ['market_activ_6_m', 'market_activ_curr_m', 'duration', \
               'prom_purch', 'avg_view_cat_per_visit', \
               'unpaid_prod_per_quart', 'service_error', \
               'pages_per_visit']  # список столбцов с выбросами
for value in column_list:  # цикл по столбцам с выбросами
    print(f'Доля выбросов в {value} в процентах: {outliers(dataset, value)}')
    # выводим долю выбросов в столбце

dataset = market_money
column_list = ['revenue']
for value in column_list:
    print(f'Доля выбросов в {value} в процентах: {outliers(dataset, value)}')
dataset = market_time
column_list = ['minutes']
for value in column_list:
    print(f'Доля выбросов в {value} в процентах: {outliers(dataset, value)}')
dataset = money
column_list = ['profit']
for value in column_list:
    print(f'Доля выбросов в {value} в процентах: {outliers(dataset, value)}')
```

Создадим функции для построения: гистограмм распределения количественных признаков (`histogram`), диаграмм распределения категориальных признаков (`categorial`) и диаграмм размаха (`whiskers`):

```python
def histogram(dataset, column):
    dataset[column].plot(kind='hist', bins=20, grid=True, \
                figsize=(5, 5), title=(f'Распределение признака "{column}"'))
    plt.xlabel(column)
    plt.ylabel('Количество')
    plt.show()
    return None

def categorial(dataset, column):
    dataset[column].value_counts().plot(
    kind = 'pie',
    ylabel = '',
    autopct = '%1.1f%%',
    figsize = (5,5),
    title = f'Процентное распределение признака "{column}"')
    plt.show()
    return None

def whiskers(dataset, column):
    dataset.boxplot(column, figsize=(5, 5))
    plt.title(f'Диаграмма размаха признака "{column}"')
    plt.ylabel('Количество')
    plt.show()
    return None
```

## Использование пайплайнов

### Подготовка данных и создание пайплайнов

Преобразуем целевой признак в числовой формат и переведём `id` клиентов в индексы:

```python
market['purch_activ'] = market['purch_activ'].apply(lambda x: 1 if \
                                            x == 'Снизилась' else 0)
market = market.set_index('id')
```

Разделим датасет на обучающую и тестовую выборки:

```python
X = market.drop('purch_activ', axis=1)
y = market['purch_activ']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    stratify=y
    )
```

Перечислим особенности данных:
- с помощью OneHotEncoder будем кодировать признаки: `allow_report` и `popular_cat`;
- с помощью OrdinalEncoder закодируем признак `service_type`;
- числовые признаки нужно масштабировать.

Создадим списки с признаками:

```python
# помещаем категориальные признаки в список:
ohe_columns = ['allow_report', 'popular_cat']
ord_columns = ['service_type']
# помещаем числовые признаки в список:
num_columns = ['market_activ_6_m', 'market_activ_curr_m', 'duration', \
               'prom_purch', 'avg_view_cat_per_visit', \
               'unpaid_prod_per_quart', 'service_error', 'pages_per_visit', \
               'revenue_two_months_ago', 'revenue_prev_month', \
               'revenue_curr_month', 'minutes_prev_month', 'minutes_curr_month']
```

В наших данных отсутствуют пропуски, тем не менее, добавим `SimpleImputer` в пайплайн.
Создадим пайплайны `ohe_pipe` и `ord_pipe` для категориальных признаков:

```python
ohe_pipe = Pipeline([
            (
            'simpleImputer_ohe',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
            ),
            (
            'ohe', OneHotEncoder(
            handle_unknown='error', sparse_output=False)
            )
            ])
ohe_pipe
```

```python
ord_pipe = Pipeline(
    [
        (
            'simple_imputer_ord_before',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',
            OrdinalEncoder()
        ),
        (
            'simple_imputer_ord_after',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)
```

Объединяем кодирование и масштабирование в один пайплайн с подготовкой данных `market_preprocessor`:

```python
market_preprocessor = ColumnTransformer(
    [
    ('ohe', ohe_pipe, ohe_columns),
    ('ord', ord_pipe, ord_columns),
    ('num', MinMaxScaler(), num_columns)
], remainder='passthrough')
market_preprocessor
```

Соберём итоговый пайплайн:

```python
pipe_final = Pipeline([
    ('preprocessor', market_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])
```

```python
# Обучим модель на тренировочной выборке с помощью пайплайна:
pipe_final.fit(X_train, y_train)

# Предсказанные значение целевого признака на тренировочной выборке:
y_pred = pipe_final.predict(X_test)
y_pred
```

### Обучение моделей

Обучим четыре модели: `KNeighborsClassifier`, `DecisionTreeClassifier`, `LogisticRegression`, `SVC`, подобрав для каждой из них как минимум один гиперпараметр

```python
param_grid = [
  # Словарь для модели KNeighborsClassifier
  {'models': [KNeighborsClassifier()],
  'models__n_neighbors': range(2,5),
  'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']},

  # Словарь для модели DecisionTreeClassifier
  {'models': [DecisionTreeClassifier()],
  'models__max_depth': range(2,5),
  'models__max_features': range(2,5),
  'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']},

  # Словарь для модели LogisticRegression
  {'models': [LogisticRegression(
      random_state=RANDOM_STATE,
      solver='liblinear',
      penalty='l1')],
  'models__C': range(2,5),
  'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']},

  # Словарь для модели SVC
  {'models': [SVC(probability=True, random_state=RANDOM_STATE, kernel='poly')],
  'models__degree': range(2,5),
  'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']}
]
```

Будем использовать метрику ROC-AUC, как наиболее подходящую для решения нашей задачи по следующим причинам:
- устойчивость к несбалансированным классам (ROC-AUC не зависит от порога классификации);
- возможность сравнения различных моделей классификации (чем ближе значение ROC-AUC к 1, тем лучше модель справляется с задачей классификации);
- инвариантность к масштабу предсказаний (ROC-AUC не зависит от масштаба вероятностей, которые генерирует модель)

### Выбираем лучшую модель, используя метрику ROC-AUC

Будем использовать рандомизированный поиск, т.к. у нас большое количество гиперпараметров и возможных комбинаций с ними:

```python
randomized_search = RandomizedSearchCV(
    pipe_final,
    param_grid,
    cv=5,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)

randomized_search.fit(X_train, y_train)
```

Выведем информацию о лучшей модели и её параметрах, а также значения метрики ROC-AUC на тренировочной и тестовой выборках:

```python
y_test_pred = randomized_search.predict_proba(X_test)[:, 1]
```

```python
print('Лучшая модель: ', randomized_search.best_estimator_)
print ('ROC-AUC лучшей модели на тренировочной выборке: ', \
       randomized_search.best_score_)
print(f'ROC-AUC на тестовой выборке: {roc_auc_score(y_test, y_test_pred)}')
```

Лучшей моделью оказалась SVC (метод опорных векторов), на тренировочной выборке она показывает значения коэффициента ROC-AUC = 0.91, на тестовой - 0.92

- проведённый анализ по группам пользователей позволяет с уверенностью порекомендовать бизнесу сосредоточиться на категории "Товары для детей" - она пользуется наибольшим спросом (примерно по 1,6 млн. руб. выручки каждый месяц), приносит максимальную выручку и именно клиентов, предпочитающих эту категорию, будет наиболее болезненно потерять;
- в то же время категория "Кухонная посуда" - совсем не так популярна (около 700 тыс. руб. выручки каждый месяц), и её ассортимент можно не увеличивать.

**Резюме по проведённой работе:** бизнесу необходимо предпринять все возможные меры по привлечению и удержанию клиентов на сайте. Увеличить ассортимент, количество страниц для просмотра, акционные товары, выгодные предложения и т.д., сосредоточившись на категории "Детские товары". Категорией "Кухонная посуда" же наоборот, можно уделить гораздо меньше внимания, несмотря на то, что она в топе рейтинга. Клиенты хотят проводить больше времени на сайте, откладывать покупки в корзину - это хороший знак. Их не беспокоит количество ошибок сайта и спам, на этом можно не заострять внимание (хотя баги стоит убрать). Возможно, не лишним будет добавить нотификации о том, точно ли клиент хочет удалить товар из корзины и не желает ли взглянуть на похожий товар.
