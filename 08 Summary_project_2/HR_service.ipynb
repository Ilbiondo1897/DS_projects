Создадим функцию для подсчёта доли выбросов в количественных столбцах всех датасетов:

```python
def outliers(dataset, column):
    q1 = dataset[column].quantile(0.25) # первый квартиль
    q3 = dataset[column].quantile(0.75)  # третий квартиль
    iqr = q3 - q1  # межквартильный размах
    quant1 = q1 - 1.5*iqr  # нижняя граница выбросов
    quant3 = q3 + 1.5*iqr  # верхняя граница выбросов
    total = dataset.shape[0]  # общее количество строк в датафрейме
    emission = dataset[(dataset[column] < quant1) | \
     (dataset[column] > quant3)].shape[0]
    # количество выбросов
    emission_quant = (emission / total) * 100  # доля выбросов в столбце
    return np.round((emission_quant), decimals=2)
    # возвращаем долю выбросов в столбце

# Столбцы с выбросами
dataset = train_jsr
column_list = ['employment_years', 'supervisor_evaluation', \
              'salary', 'job_satisfaction_rate']  # список столбцов с выбросами
for value in column_list:  # цикл по столбцам с выбросами
    print(f'Доля выбросов в {value} в процентах: {outliers(dataset, value)}')
    # выводим долю выбросов в столбце

dataset = test_features
column_list = ['employment_years', 'supervisor_evaluation', 'salary']
for value in column_list:
    print(f'Доля выбросов в {value} в процентах: {outliers(dataset, value)}')

dataset = test_target_jsr
column_list = ['job_satisfaction_rate']
for value in column_list:
    print(f'Доля выбросов в {value} в процентах: {outliers(dataset, value)}')
```

Доли выбросов в `supervisor_evaluation` и `salary` незначительны, тем не менее, изучим диаграммы размаха для этих столбцов более внимательно.

```python
# Гистограмма распределения количественных признаков
def histogram(dataset, column):
    dataset[column].plot(kind='hist', bins=20, grid=True, \
                figsize=(5, 5), title=(f'Распределение признака "{column}"'))
    plt.xlabel(column)
    plt.ylabel('Количество')
    plt.show()
    return None

# Диаграмма распределения категориальных признаков
def categorial(dataset, column):
    dataset[column].value_counts().plot(
    kind = 'pie',
    ylabel = '',
    autopct = '%1.1f%%',
    figsize = (5,5),
    title = f'Процентное распределение признака "{column}"')
    plt.show()
    return None

# Диаграмма размаха
def whiskers(dataset, column):
    dataset.boxplot(column, figsize=(5, 5))
    plt.title(f'Диаграмма размаха признака "{column}"')
    plt.ylabel('Количество')
    plt.show()
    return None
```

### Подготовка данных

На основе проведённого ранее анализа, определим, какие признаки и каким образом будем кодировать и масштабировать:

```python
# поместим категориальные признаки в списки:
ohe_cols = ['dept', 'last_year_promo', 'last_year_violations']
ord_cols = ['level', 'workload']
# поместим числовые признаки в список:
num_cols = ['employment_years', 'supervisor_evaluation', 'salary']
```

Объединяем кодирование и масштабирование в один пайплайн с подготовкой данных `data_preprocessor`:

```python
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_cols),
        ('ord', ord_pipe, ord_cols),
        ('num', MinMaxScaler(), num_cols)
    ],
    remainder='passthrough'
)
data_preprocessor
```

### Обучение моделей

Объединим датасеты `test_features` и `test_target_jsr` по общему столбцу `id`:

```python
test_features = test_features.merge(test_target_jsr, on='id', how='left')
test_features.head()
```

Переименуем для удобства тренировочные и тестовые выборки в более привычные обозначения:

```python
X_train = train_jsr.drop('job_satisfaction_rate', axis=1)
y_train = train_jsr['job_satisfaction_rate']
X_test = test_features.drop('job_satisfaction_rate', axis=1)
y_test = test_features['job_satisfaction_rate']
```

Соберём итоговый пайплайн:

```python
pipe_final = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', LinearRegression())
    ]
)
pipe_final
```

Обучим две модели: `линейная регрессия` и `дерево решений`. Подберём гиперпараметры `max_depth`, `min_samples_split` и `max_features` для второй из них:

```python
param_grid = [
    {
        'models': [LinearRegression()],
        'preprocessor__num': [MinMaxScaler(), StandardScaler(), \
                              'passthrough']
    },
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(1, 22),
        'models__min_samples_split': range(1, 22),
        'models__max_features': range(1, 22),
        'preprocessor__num': [MinMaxScaler(), StandardScaler(), \
                              'passthrough']
    }
]
```

Создадим новую метрику - `SMAPE` (symmetric mean absolute percentage error, «симметричное среднее абсолютное процентное отклонение»). Напишем функцию, которая принимает на вход массивы NumPy в Pandas и возвращает значение метрики SMAPE:

```python
def smape(y_true, y_pred):
    return 1/len(y_true) * np.sum(2 * \
    np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred)) * 100)
```

```python
# Создание метрики
smape_scorer = make_scorer(smape, greater_is_better=False)
```

Будем использовать рандомизированный поиск, т.к. у нас большое количество гиперпараметров и возможных комбинаций с ними:

```python
randomized_search = RandomizedSearchCV(
    pipe_final,
    param_grid,
    scoring=smape_scorer,
    cv=5,
    n_jobs=-1,
    random_state=RANDOM_STATE
)
randomized_search.fit(X_train, y_train)
```

```python
print(f'Параметры лучшей модели:\n{randomized_search.best_params_}')
print ('Метрика лучшей модели на кросс-валидации:', \
       round(randomized_search.best_score_, 2))
```

Победила модель DecisionTreeRegressor с гиперпараметрами: `min_samples_split` = 9, `max_features` = 19, `max_depth` = 18. Метрика лучшей модели на кросс-валидации: 15.19

Выведем результаты оценки топ-10 лучших моделей:

```python
top_results = pd.DataFrame(randomized_search.cv_results_) \
                  .loc[:, ['params', 'mean_test_score', 'rank_test_score']]
top_results.sort_values(by='rank_test_score').head(10)
```

Критерием успеха, согласно поставленной нам задаче, будет являться полученное значение SMAPE ≤ 15 на тестовой выборке. Убедимся в том, что мы выполнили условие:

```python
best_model = randomized_search.best_estimator_
print(f'SMAPE на тестовой выборке: \
{round(smape(y_test, best_model.predict(X_test)), 2)}')
```

Значение метрики SMAPE на тестовой выборке равно 13.72, что соответствует поставленному условию.

- модель "дерево решений" справилась лучше модели "линейная регрессия", т.к. у нас много нелинейности в данных (в чём мы убедились ранее, построив матрицу корреляции).

### Подготовка данных

```python
# поместим категориальные признаки в списки:
ohe_cols = ['dept', 'last_year_promo', 'last_year_violations']
ord_cols = ['level', 'workload']
# поместим числовые признаки в список:
num_cols = ['employment_years', 'supervisor_evaluation', \
            'salary', 'job_satisfaction_rate']
```

Объединяем кодирование и масштабирование в один пайплайн с подготовкой данных `data_preprocessor`:

```python
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_cols),
        ('ord', ord_pipe, ord_cols),
        ('num', MinMaxScaler(), num_cols)
    ],
    remainder='passthrough'
)
data_preprocessor
```

Выводы по разделу:

- разложили признаки в списки по способам их кодирования и масштабирования;
- создали единый пайплайн с подготовкой данных.

### Обучение модели

Переименуем для удобства тренировочные и тестовые выборки в более привычные обозначения:

```python
X_train = train_quit.drop('quit', axis=1)
y_train = train_quit['quit']

X_test = test_features.drop('quit', axis=1)
y_test = test_features['quit']
```

Проверим, что разделение на выборки прошло успешно:

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_test.head()
```

```python
y_test.head()
```

Соберём итоговый пайплайн `pipe_final_new`:

```python
pipe_final_new = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', LogisticRegression())
    ]
)
pipe_final_new
```

Обучим четыре модели и подберём для них гиперпараметры:

```python
param_grid_new = [
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(1, 20),
        'preprocessor__num': [MinMaxScaler(), StandardScaler()]
    },
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(1, 11),
        'models__max_features': range(1, 11),
        'models__min_samples_leaf': range(1, 11),
        'preprocessor__num': [MinMaxScaler(), StandardScaler(), \
                              'passthrough']
    },
    {
        'models': [SVC(probability=True, random_state=RANDOM_STATE)],
        'models__C': range(1, 10),
        'models__kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
        'preprocessor__num': [MinMaxScaler(), StandardScaler(), \
                              'passthrough']
    },
    {
        'models': [LogisticRegression(random_state=RANDOM_STATE)],
        'models__C': range(1, 20),
        'models__penalty': ['l1', 'l2', 'elasticnet', 'none'],
        'preprocessor__num': [MinMaxScaler(), StandardScaler(), \
                              'passthrough']
    }
]
```

Будем использовать рандомизированный поиск, т.к. у нас большое количество гиперпараметров и возможных комбинаций с ними:

```python
randomized_search_new = RandomizedSearchCV(
    pipe_final_new,
    param_grid_new,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)

randomized_search_new.fit(X_train, y_train)
```

Выведем лучшую модель:

```python
print(f'Параметры лучшей модели:\n{randomized_search_new.best_params_}')
print ('Метрика лучшей модели на кросс-валидации:', \
       round(randomized_search_new.best_score_, 2))
```

Победила модель SVC с гиперпараметрами: `kernel` = "rbf", `C` = 5. Метрика лучшей модели на кросс-валидации: 0.93

Выведем результаты оценки топ-10 лучших моделей:

```python
top_results_new = pd.DataFrame(randomized_search_new.cv_results_) \
                  .loc[:, ['params', 'mean_test_score', 'rank_test_score']]
top_results_new.sort_values(by='rank_test_score').head(10)
```

```python
best_model_new = randomized_search_new.best_estimator_
y_pred = best_model_new.predict_proba(X_test)[:, 1]
roc_auc_score(y_test, y_pred)
```
